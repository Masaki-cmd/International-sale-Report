### 1. データ型の変換

欠損値処理の前に、`PCS`、`RATE`、`GROSS AMT`列が現在`object`型ですが、数値として扱われるべきです。また、`DATE`列も日付型に変換する必要があります。これらの列を適切な型に変換し、変換できない値は`NaN`（欠損値）として扱います。

```python
import numpy as np

# 'PCS', 'RATE', 'GROSS AMT' を数値型に変換
# 変換できない値は NaN にする (errors='coerce')
df['PCS'] = pd.to_numeric(df['PCS'], errors='coerce')
df['RATE'] = pd.to_numeric(df['RATE'], errors='coerce')
df['GROSS AMT'] = pd.to_numeric(df['GROSS AMT'], errors='coerce')

# 'DATE' 列を日付型に変換
df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce', dayfirst=True)

print('\n--- After Type Conversion ---')
df.info()
print('\n--- Missing Values After Type Conversion ---')
display(df.isnull().sum())
```

### 2. 欠損値を含む行の削除 (`dropna()`)

最も簡単な方法は、欠損値を含む行全体を削除することです。データ量が非常に多く、欠損値の割合が少ない場合に適しています。しかし、欠損値が多い場合や特定の列に集中している場合は、多くの有用なデータが失われる可能性があります。

ここでは、`DATE`、`Months`、`CUSTOMER`、`Style`、`Size`、`PCS`、`RATE`、`GROSS AMT`のいずれかに欠損値がある行を削除する例を示します。`SKU`は欠損値が多いため、ここでは対象外として、別の補完方法を検討します。

```python
# 欠損値処理のためにデータフレームをコピー
df_dropped = df.copy()

# 特定の列のサブセットで欠損値がある行を削除
columns_to_check_for_na = ['DATE', 'Months', 'CUSTOMER', 'Style', 'Size', 'PCS', 'RATE', 'GROSS AMT']
df_dropped.dropna(subset=columns_to_check_for_na, inplace=True)

print(f"--- Original DataFrame shape: {df.shape} ---")
print(f"--- DataFrame shape after dropping rows with NA in specified columns: {df_dropped.shape} ---")
print('\n--- Missing Values After Dropping Rows ---')
display(df_dropped.isnull().sum())
```

### 3. 欠損値の補完 (Imputation)

データを失うことなく欠損値を埋める方法です。列のデータ型と性質に応じて、様々な補完戦略があります。

#### 数値型列の補完

`PCS`、`RATE`、`GROSS AMT`のような数値型データの場合、平均値、中央値、または特定の値（例: 0）で補完することが一般的です。

*   **平均値 (Mean)**: データの分布が正規分布に近い場合や、外れ値の影響が少ない場合に適しています。
*   **中央値 (Median)**: 外れ値の影響を受けにくい補完方法で、データに偏りがある場合に有効です。

#### カテゴリ型列（オブジェクト型）の補完

`Months`、`CUSTOMER`、`Style`、`SKU`、`Size`のようなカテゴリ型データの場合、最頻値（Mode）で補完するか、「Unknown」のような新しいカテゴリで補完することが一般的です。

*   **最頻値 (Mode)**: 最も頻繁に出現するカテゴリで補完します。
*   **固定値/新しいカテゴリ**: 欠損が特定のパターンによって発生している可能性がある場合や、欠損自体が意味を持つ場合に「不明」などの新しいカテゴリとして扱います。

ここでは、`df`の元のデータフレームに補完を適用します。

```python
# 数値型列の補完 (中央値を使用)
for col in ['PCS', 'RATE', 'GROSS AMT']:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"Column '{col}' imputed with median: {median_val}")

# カテゴリ型列の補完
# Months, CUSTOMER, Style, Size は最頻値で補完
for col in ['Months', 'CUSTOMER', 'Style', 'Size']:
    if df[col].isnull().any():
        mode_val = df[col].mode()[0] # mode() はSeriesを返すため、[0]で値を取得
        df[col].fillna(mode_val, inplace=True)
        print(f"Column '{col}' imputed with mode: {mode_val}")

# SKU は欠損値が多いため、「Unknown」で補完
if df['SKU'].isnull().any():
    df['SKU'].fillna('Unknown', inplace=True)
    print(f"Column 'SKU' imputed with 'Unknown'")

# DATE列の補完（例: 最頻値で補完。ただし、日付データは文脈によって補完方法が大きく異なる）
if df['DATE'].isnull().any():
    # 最も頻繁な日付で補完
    mode_date = df['DATE'].mode()[0]
    df['DATE'].fillna(mode_date, inplace=True)
    print(f"Column 'DATE' imputed with mode: {mode_date}")

print('\n--- Missing Values After Imputation ---')
display(df.isnull().sum())
```

### 4. 売上金額 (`GROSS AMT`) の月別推移の可視化

`DATE`列から月を抽出し、月ごとの合計売上金額を計算して、その推移を折れ線グラフで可視化します。

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 'DATE' 列が datetime 型であることを確認し、年と月を抽出
df['YearMonth'] = df['DATE'].dt.to_period('M')

# 月ごとに GROSS AMT を集計
monthly_sales = df.groupby('YearMonth')['GROSS AMT'].sum().reset_index()

# YearMonth を datetime オブジェクトに変換してプロット用に準備
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()

# 月別売上推移を可視化
plt.figure(figsize=(15, 7))
sns.lineplot(x='YearMonth', y='GROSS AMT', data=monthly_sales)
plt.title('Monthly Sales Trend (GROSS AMT)')
plt.xlabel('Month')
plt.ylabel('Total Gross Amount')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

    *   `DATE`列の書式を統一し、日付への変換エラーを防ぐため`format`引数を指定した変換を推奨します。

    *   非製品のスタイル（例: `SHIPPING`, `LABEL CHARGE`, `TAGS`）は、製品売上分析から除外するか、別途「サービス費」などのカテゴリに分類し直すことを検討します。

    *   売上ゼロのスタイルが製品である場合は、その原因（プロモーション不足、価格、競合、需要の欠如など）を詳細に調査し、改善策を立案します。
    *   製品でない場合は、データ集計の段階で適切に除外されるよう、データの定義と処理フローを見直します。

#### 5. 期待される効果

#### 12.2. 「SET268」の関連商品（同時購入）分析
「SET268」が購入された際に、同時に購入されることが多い他のスタイルを特定します。これはクロスセル戦略のヒントとなります。

```python
# --- 関連商品（同時購入）分析 ---
print(f"\n--- Co-purchased Items Analysis for Style: {current_top_style} ---")

# current_top_style を購入した顧客と日付のペアを特定
co_purchase_candidates = style_transactions[['DATE', 'CUSTOMER']].drop_duplicates()

# これらの顧客と日付のペアに対応する全てのスタイルを抽出
# 同じ日付、同じ顧客による購入を「同時購入」とみなす
all_styles_in_co_purchases = df.merge(
    co_purchase_candidates,
    on=['DATE', 'CUSTOMER'],
    how='inner'
)

# current_top_style 以外のスタイルに絞る
related_styles = all_styles_in_co_purchases[all_styles_in_co_purchases['Style'] != current_top_style]

# 関連スタイルの出現回数をカウント
co_purchase_counts = related_styles['Style'].value_counts()

print(f"\nTop 10 co-purchased styles with '{current_top_style}':")
display(co_purchase_counts.head(10))

# 可視化
plt.figure(figsize=(12, 6))
sns.barplot(x=co_purchase_counts.head(10).index, y=co_purchase_counts.head(10).values, hue=co_purchase_counts.head(10).index, palette='magma', legend=False)
plt.title(f'Top 10 Co-purchased Styles with {current_top_style}')
plt.xlabel('Related Style')
plt.ylabel('Number of Co-purchases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\n--- 次のステップ ---")
print(f"上記の分析を、他の上位スタイル（SET268以外のJ0277, J0242, J0012, SET252）についても繰り返すことで、各スタイルの成功要因やマーケティング戦略のヒントを得ることができます。")
```
